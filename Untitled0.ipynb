```python
# Que tecnicas podrias aplicar para imputar los datos faltantes en 'PIB' y 'Poblacion'
import pandas as pd
data = {
    'Pais': ['Mexico', 'Colombia','Argentina', 'Peru', 'Chile'],
    'PIB': [1200, 950, None, 800, 850],
    'Poblacion': [128, None, 45, 33, 19]
}

df = pd.DataFrame(data)
```

```python
# Respuesta a - tecnicas podrias aplicar para imputar los datos faltantes en 'PIB' y 'Poblacion'
df['PIB'] = df['PIB'].fillna(df['PIB'].mean())
df['Poblacion'] = df['Poblacion'].fillna(df['Poblacion'].mean())
df
```

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd

html = urlopen('https://es.wikipedia.org/wiki/Anexo:%C3%93scar_al_mejor_actor#A%C3%B1os_2020')
bs = BeautifulSoup(html.read(), 'html.parser')

all_name_mas = bs.find_all('td')
names_mas = []

for name in all_name_mas:
  # Use find on the current 'name' element (a Tag object) instead of the ResultSet
  tg = name.find('b')
  if tg and tg.find('a'): #also ensure the 'a' tag is actually present
    names_mas.append(name.text.strip()) # Extract and clean the text content

df = pd.DataFrame({'actores': names_mas})
df
```

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://es.wikipedia.org/wiki/Anexo:%C3%93scar_a_la_mejor_pel%C3%ADcula'
html = urlopen(url)
bs = BeautifulSoup(html.read(), 'html.parser')

all_name_mov = bs.find_all('td', {'style': 'background:#; font-size:pt;'})
names_mov = []

for name in all_name_mov:
    tg = name.find('i')
    if tg and tg.find('a'):
        names_mov.append(tg.find('a').text.strip())

df = pd.DataFrame({'peliculas': names_mov})
df
```
